```python
import pandas as pd

# Creating the first DataFrame for employee information
emp = pd.DataFrame({
    'Emp_ID': [101, 102, 103, 104, 105],
    'Ename': ['Swapnil', 'Naresh', 'Prakash', 'Rohan', 'Swanand'],
    'Dept_ID': [11, 22, 22, 44, 66]
})

# Displaying the first DataFrame
print(emp)

# Creating the second DataFrame for department information
dept = pd.DataFrame({
    'Dept_ID': [11, 22, 33, 44],
    'Dname': ['Admin', 'Software Development', 'HR', 'Sales'],
    'Location': ['D0', 'D1', 'D1', 'D3']
})

# Displaying the second DataFrame
print(dept)
```

```python
x1=pd.merge(dept,emp,how='inner',on='Dept_ID')
x1
```

```python
x1=pd.merge(dept,emp,how='right',on='Dept_ID')
x1
```

```python
x1=pd.merge(dept,emp,how='left',on='Dept_ID')
x1
```

```python
x1=pd.merge(dept,emp,how='outer',on='Dept_ID')
x1
```
