# interpolation

```python
import pandas as pd
import numpy as np

data = {
    'States': ['Maharashtra', 'Karnataka', 'Maharashtra', 'Gujarat', np.nan,
               'West Bengal', 'Rajasthan', 'Gujarat'],
    'States_id': [10, 20, 10, 40, np.nan, 50, 60, 40],
    'Dates': pd.to_datetime(['2024-01-01', '2024-01-15', '2024-02-18', '2024-02-15',
                             '2024-03-01', '2024-03-15', '2024-03-20', '2024-03-31']),
    'Sales in Jan': [5000, 5200, np.nan, 5800, 6000, np.nan, 6400, 6600],
    'Sales in Feb': [7000, np.nan, 7500, np.nan, 8000, 8200, np.nan, 8600],
    'Sales in March': [8800, 9200, 9400, np.nan, 9800, np.nan, 10200, 10400]
}

df = pd.DataFrame(data)
df
```

```python
df.interpolate()
```

```python
df.set_index('Dates')
```

```python
df.set_index('Dates', inplace=True)
df.interpolate(method='time')
```

```python
df.reset_index(inplace=True)
df
```

```python
df['States']=df['States'].bfill()
df
```

```python
df.reset_index(inplace=True)
df['States_id'] = df['States_id'].bfill()
df['States_id'] = df['States_id'].astype(int)
df
```

```python
df.set_index('States_id', inplace=True)
df
```

```python
df[['Sales in Jan', 'Sales in Feb', 'Sales in March']] = df[['Sales in Jan', 'Sales in Feb', 'Sales in March']].interpolate(method='index')
df
```

# Pivot & Pivot_tables

```python
import pandas as pd

df = pd.DataFrame({
    'State': ['Gujarat', 'Gujarat', 'Maharashtra', 'Maharashtra', 'Gujarat'],
    'Month': ['Jan', 'Feb', 'Jan', 'Feb', 'Jan'],
    'Sales': [1000, 1200, 1500, 1800, 900]
})

df
```

```python
pivot_df=df.pivot_table(index='State', columns='Month', values='Sales')
pivot_df
```

```python
import pandas as pd

# Creating a DataFrame
df = pd.DataFrame({
    'State': ['Gujarat', 'Uttar Pradesh', 'Maharashtra', 'Goa', 'Bihar'],
    'Month': ['Mar', 'Feb', 'Mar', 'Feb', 'Mar'],
    'Sales': [1000, 1200, 1500, 1800, 900],
    'Profit': [34.12, 45.32, 55.67, 59.45, 24.45]
})

# Display the DataFrame
print(df)
```

# Encoding Categorical Data to Numeric

```python
df=pd.DataFrame({'Employee':['Alice','Bob','Charlie','David'],'Salary':[90000,50000,75000,45000]})

def categorical_salary(Salary):
    if Salary > 80000:
        return 'High'
    elif Salary > 50000:
        return 'Medium'
    else:
        return 'Low'
    
#df['Salary Category'] = df['Salary'].apply(categorical_salary)
#you can also use apply

df['Salary Category'] = df['Salary'].map(categorical_salary)
df
```

```python
import pandas as pd

df = pd.DataFrame({
    'Employee': ['Alice', 'Bob', 'Charlie', 'David'],
    'Department': ['HR', 'IT', 'Finance', 'IT']
})

# Mapping department names to numerical codes
department_map = {'HR': 1, 'IT': 2, 'Finance': 3}
df['Department Code'] = df['Department'].map(department_map)

df
```
