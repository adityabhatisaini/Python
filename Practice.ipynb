```python
import numpy as np
import pandas as pd

# Generate normally distributed data
np.random.seed(42)
data = {'Scores': np.random.normal(loc=70, scale=10, size=100)}  # Mean = 70, Std Dev = 10

df = pd.DataFrame(data)

# Introduce some extreme outliers
df.loc[5, 'Scores'] = 150
df.loc[20, 'Scores'] = 5
df.loc[50, 'Scores'] = 160
df
```

```python
data = {('Region', ''): ['North', 'North', 'North','South', 'South','South'],
        ('City', ''): ['Delhi', 'Lucknow', 'Gandhinagar','Chennai', 'Bangalore','Hyderabad'],
        ('Sales', 'Q1'): [100, 200, None, 300,250,170],
        ('Sales', 'Q2'): [150, None, 250, 350,300,240]}

df = pd.DataFrame(data)
df.columns = pd.MultiIndex.from_tuples(df.columns)
df.set_index([('Region', ''), ('City', '')], inplace=True)
df
```

```python
import numpy as np
import pandas as pd

# Generate normally distributed exam scores
np.random.seed(42)
data = {'Exam Scores': np.random.normal(loc=75, scale=12, size=100)}  # Mean = 75, Std Dev = 12

df = pd.DataFrame(data)

# Introduce some extreme outliers
df.loc[10, 'Exam Scores'] = 150
df.loc[25, 'Exam Scores'] = 10
df.loc[60, 'Exam Scores'] = 5

df
```
