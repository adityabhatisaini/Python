```python
new_df = pd.DataFrame()
new_df['Platform'] = df['publisher_platform']
new_df['Ad Date'] = df['ad_date']
new_df['Total Spends'] = df['spend']
new_df['Total Impressions'] = df['impressions']
new_df['Total Clicks'] = df['clicks']
new_df['Total Unique Clicks'] = df['unique_clicks']
new_df
```

```python
new_df['CPM'] = (df['spend'] * 1000) / df['impressions']
new_df['CTR'] = (df['clicks'] / df['impressions']) * 100
new_df['Unique CTR'] = (df['unique_clicks'] / df['impressions']) * 100
new_df.head(5)
```
