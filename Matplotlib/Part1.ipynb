## Histogram

```python
Python=np.random.randint(100,1000,(100))
Java=np.random.randint(100,1000,(100))
bins=[100,200,300,400,500,600,700,800,900,1000]

plt.hist(Python,bins,rwidth=0.8)
plt.xlim(100,1000)
plt.ylim(0,20)
plt.title('Python&Java Histogram')
plt.xlabel('Number of bins')
plt.ylabel('Number of Students')

plt.show()
```

```python
Python=np.random.randint(100,1000,(100))
Java=np.random.randint(100,1000,(100))
bins=[100,200,300,400,500,600,700,800,900,1000]
plt.figure(figsize=(10,6))
plt.hist([Java,Python],bins,rwidth=0.8,color=['b','r'])
plt.xlim(100,1000)
plt.ylim(0,20)
plt.title('Python&Java Histogram')
plt.xlabel('Number of bins')
plt.ylabel('Number of Students')
plt.legend(['Python','Java'],loc='upper right')
plt.text(400,15,'Python Hist')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random age data for 100 students
ML_students = np.random.randint(18, 45, 100)
bins = [15, 20, 25, 30, 35, 40, 45]

# Create a figure with specified size
plt.figure(figsize=(20, 10))

# Plot histogram
counts, bins, _ = plt.hist(ML_students, bins, rwidth=0.8, color='skyblue')

# Annotate each bar with the count of students
for i, count in enumerate(counts):
    plt.text(bins[i] + (bins[i + 1] - bins[i]) / 2, count + 1, int(count), ha='center')

# Customize the plot
plt.title('Age Distribution of ML Students')
plt.xlabel('Age of Students')
plt.ylabel('Number of Students')
plt.grid(axis='y', linestyle='--')
plt.ylim(0, 30)

# Show the plot
plt.show()
```

```python
classes = ['python', 'Numpy', 'Pandas', 'ML', 'Excel']
C_1 = [45, 34, 39, 23, 67]
C_2 = [34, 56, 44, 23, 54]
C_3 = [22, 33, 44, 66, 23]
width = 0.2
class_length = np.arange(len(classes))
plt.bar(class_length, C_1,width,label='Class 1')
plt.bar(class_length+width, C_2,width,label='Class 2')
plt.bar(class_length+2*width, C_3,width,label='Class 3')
plt.legend(loc='upper left')
plt.xticks(class_length+width,classes)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    "Category": ["Electronics", "Electronics", "Furniture", "Furniture", "Clothing", "Clothing"],
    "Region": ["North", "South", "North", "South", "North", "South"],
    "Sales": [1000, 500, 700, 200, 300, 400]
}

df = pd.DataFrame(data)
grouped_data = df.groupby(["Category", "Region"])["Sales"].sum()

unstacked_data = grouped_data.unstack()

unstacked_data.plot(kind="bar", figsize=(8, 6))

plt.title("Sales by Category and Region")
plt.xlabel("Category")
plt.ylabel("Sales")
plt.legend(title="Region")
plt.grid(axis='y', linestyle="--", alpha=0.7)
plt.tight_layout()
plt.show()
```
