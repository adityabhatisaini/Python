```python

labels = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
values = [10, 14, 12, 15, 18, 20, 25, 22, 19, 17, 13, 11]
```

```python
plt.plot(labels, values,marker='o',color='b',linestyle='--',linewidth=1,markersize=10)
plt.xticks(rotation=30)
plt.title('Monthly Distribution')
plt.xlabel('Months')
plt.ylabel('Max Temperature (C)')
plt.xlim(0, 12)
plt.ylim(0, 30)

plt.show()
```

```python
import matplotlib.pyplot as plt

# Create a new figure with a specific background color
fig = plt.figure(facecolor='lightblue')

# Add a subplot
ax = fig.add_subplot(111)

# Set the background color of the axes
ax.set_facecolor('lightgreen')

# Plot some data
ax.plot([1, 2, 3], [4, 5, 6])

# Show the plot
plt.show()
```
