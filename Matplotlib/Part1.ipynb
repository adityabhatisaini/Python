```python

labels = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
values = [10, 14, 12, 15, 18, 20, 25, 22, 19, 17, 13, 11]
```

```python
plt.plot(labels, values,marker='o',color='b',linestyle='--',linewidth=1,markersize=10)
plt.xticks(rotation=30)
plt.title('Monthly Distribution')
plt.xlabel('Months')
plt.ylabel('Max Temperature (°C)')
plt.xlim(0, 12)
plt.ylim(0, 30)

plt.show()
```

```python
df = pd.DataFrame({'Year': range(2009, 2017), 'Avg House Prices in Nagpur': [25e6, 23e6, 20e6, 30e6, 32e6, 28e6, 35e6, 40e6]})
df
```

```python
plt.plot(df['Year'],df['Avg House Prices in Nagpur'],color='#FFD700',linestyle='--',linewidth=1,markersize=10,label='Avg House Prices in Nagpur')
plt.xlabel('Avg House Prices in Nagpur')
plt.ylabel('Year')
plt.xlim(2009,2017)
plt.ylim(20e6,45e6)
plt.legend(loc='lower right')
plt.title('Average House Prices in Nagpur from 2009 to 2016')
plt.show()
```
